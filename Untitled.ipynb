```python
import os
import pandas as pd

# 定义根目录
# root_dir = '/data2/wangzhongren/taolin_project/FuxiCTR/model_zoo/DeepFM/DeepFM_torch/checkpoints/'
# root_dir = '/data2/wangzhongren/taolin_project/FuxiCTR/model_zoo/DCNv2/checkpoints/'
# root_dir = '/data2/wangzhongren/taolin_project/FuxiCTR/model_zoo/AutoInt/checkpoints/'
root_dir = '/data2/wangzhongren/taolin_project/FuxiCTR/model_zoo/DeepIM/checkpoints/'

# 定义方法和尺度
methods = ['moe', 'rq', 'me']
scales = [1, 3, 7]
seeds = ['seed_1027', 'seed_20', 'seed_201', 'seed_2024', 'seed_2333']

# 初始化一个空的列表来存储数据
data = []

# 遍历方法和尺度文件夹
for method in methods:
    for scale in scales:
        for seed in seeds:
            log_file_path = os.path.join(root_dir, f'split-{method}-{scale}', seed, f'split-{method}-{scale}.log')
            if os.path.exists(log_file_path):
                with open(log_file_path, 'r') as file:
                    lines = file.readlines()
                    if lines:
                        last_line = lines[-1].strip()
                        if 'AUC' in last_line and 'logloss' in last_line:
                            parts = last_line.split(' ')
                            auc = float(parts[parts.index('AUC:') + 1])
                            logloss = float(parts[parts.index('logloss:') + 1])
                            data.append([method, scale, seed, auc, logloss])

# 处理 split-base 文件夹
base_method = 'base'
for seed in seeds:
    log_file_path = os.path.join(root_dir, f'split-{base_method}', seed, f'split-{base_method}.log')
    if os.path.exists(log_file_path):
        with open(log_file_path, 'r') as file:
            lines = file.readlines()
            if lines:
                last_line = lines[-1].strip()
                if 'AUC' in last_line and 'logloss' in last_line:
                    parts = last_line.split(' ')
                    auc = float(parts[parts.index('AUC:') + 1])
                    logloss = float(parts[parts.index('logloss:') + 1])
                    data.append([base_method, 1, seed, auc, logloss])  # Scale 对于 base 方法设为 None

# 将数据转换为 DataFrame
df = pd.DataFrame(data, columns=['Method', 'Scale', 'Seed', 'AUC', 'Logloss'])

# 打印整理后的 DataFrame
print(df)
```

```python
summary_df = df.groupby(['Method', 'Scale']).agg(
    AUC_mean=('AUC', 'mean'),
    AUC_std=('AUC', 'std'),
    Logloss_mean=('Logloss', 'mean'),
    Logloss_std=('Logloss', 'std')
).reset_index()
```
